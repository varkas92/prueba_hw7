```python
obs_data = np.loadtxt("lotka_volterra_obs.dat")
t_obs = obs_data[:, 0]
x_obs = obs_data[:, 1]
y_obs = obs_data[:, 2]

n = 96

x = np.ones(n)
y = np.ones(n)

x[0] = x_obs[0]
y[0] = y_obs[0]

n_steps = 20000
n_burn = 5000
```

```python
figsize(15, 15)
plt.scatter(t_obs, x_obs, c = "r")
plt.scatter(t_obs, y_obs, c = "g")
title("Datos iniciales")
```

```python
def likelihood(x_obs, x_model, y_obs, y_model):
    
    chi_squared_x = (1.0 / 2.0) * sum((x_obs - x_model) ** 2)
    chi_squared_y = (1.0 / 2.0) * sum((y_obs - y_model) ** 2)
    
    return chi_squared_x + chi_squared_y
```

```python
def x_prima(x, y, alpha, beta):
    
    return x * (alpha - beta * y)

def y_prima(x, y, gamma, delta):
    
    return -y * (gamma - delta * x)

def RK4(ti, tf, xi, yi, alpha, beta, gamma, delta):
    
    h = tf - ti
    
    k_1_x = x_prima(xi, yi, alpha, beta)
    k_1_y = y_prima(xi, yi, gamma, delta)
    
    #Paso 1
    t1 = ti + (h / 2.0)
    
    x1 = xi + (h / 2.0) * k_1_x
    y1 = yi + (h / 2.0) * k_1_y
    
    k_2_x = x_prima(x1, y1, alpha, beta)
    k_2_y = y_prima(x1, y1, gamma, delta)
    
    #Paso 2
    t2 = ti + (h / 2.0)
    
    x2 = xi + (h / 2.0) * k_2_x
    y2 = yi + (h / 2.0) * k_2_y
    
    k_3_x = x_prima(x2, y2, alpha, beta)
    k_3_y = y_prima(x2, y2, gamma, delta)
    
    #Paso 3
    t3 = ti + h
    
    x3 = xi + h * k_3_x
    y3 = yi + h * k_3_y
    
    k_4_x = x_prima(x3, y3, alpha, beta)
    k_4_y = y_prima(x3, y3, gamma, delta)
    
    #Paso 4
    av_k_x = (1.0 / 6.0) * (k_1_x + 2.0 * k_2_x + 2.0 * k_3_x + k_4_x)
    av_k_y = (1.0 / 6.0) * (k_1_y + 2.0 * k_2_y + 2.0 * k_3_y + k_4_y)
    
    x_new = xi + h * av_k_x
    y_new = yi + h * av_k_y
    
    #xi = xi + h * av_k_x
    #yi = yi + h * av_k_y
    
    return x_new, y_new
```

```python
#Runge-Kutta cuarto orden
#def my_model(t_obs, x, y, alpha, beta, gamma, delta):
def my_model(alpha, beta, gamma, delta):
    
    n = 96

    for i in range(1, n):
        
        #x_new = x[i - 1]
        #y_new = y[i - 1]
    
        x[i], y[i] = RK4(t_obs[i - 1], t_obs[i], x[i - 1], y[i - 1], alpha, beta, gamma, delta)
        
        #RK4(t_obs[i - 1], t_obs[i], x_new, y_new, alpha, beta, gamma, delta)
        
        #x[i] = x_new
        #y[i] = y_new
        
    return x, y
```

```python
alpha_walk = np.ones(n_steps)
beta_walk = np.ones(n_steps)
gamma_walk = np.ones(n_steps)
delta_walk = np.ones(n_steps)

l_walk = np.ones(n_steps)

alpha_walk[0] = random.random()
beta_walk[0] = random.random()
gamma_walk[0] = random.random()
delta_walk[0] = random.random()

x_init, y_init = my_model(alpha_walk[0], beta_walk[0], gamma_walk[0], delta_walk[0])
l_walk[0] = likelihood(x_obs, x_init, y_obs, y_init)

print alpha_walk[0]
print beta_walk[0]
print gamma_walk[0]
print l_walk[0]
print x[0], y[0]
```

```python
for i in range(0, (n_steps - 1)):
    
    alpha_prime = np.random.normal(alpha_walk[i], 0.1) 
    beta_prime = np.random.normal(beta_walk[i], 0.1)
    gamma_prime = np.random.normal(gamma_walk[i], 0.1)
    delta_prime = np.random.normal(delta_walk[i], 0.1) 

    x_init, y_init = my_model(alpha_walk[i], beta_walk[i], gamma_walk[i], delta_walk[i])
    x_prime, y_prime = my_model(alpha_prime, beta_prime, gamma_prime, delta_prime)
    
    l_init = likelihood(x_obs, x_init, y_obs, y_init)
    l_prime = likelihood(x_obs, x_prime, y_obs, y_prime)
    
    gam = -l_prime + l_init
    
    if(gam >= 0.0):
        
        alpha_walk[i + 1] = alpha_prime
        beta_walk[i + 1] = beta_prime
        gamma_walk[i + 1] = gamma_prime
        delta_walk[i + 1] = delta_prime
        l_walk[i + 1] = l_prime
        
    else:
        
        bet = log(random.random())
        #alp = exp(gam)
        
        if(bet <= gam):
            
            alpha_walk[i + 1] = alpha_prime
            beta_walk[i + 1] = beta_prime
            gamma_walk[i + 1] = gamma_prime
            delta_walk[i + 1] = delta_prime
            l_walk[i + 1] = l_prime
               
        else:
            
            alpha_walk[i + 1] = alpha_walk[i]
            beta_walk[i + 1] = beta_walk[i]
            gamma_walk[i + 1] = gamma_walk[i]
            l_walk[i + 1] = l_init
```
